# Data Visualization with Matplotlib:


```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
dataset : DataFrame = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

```python
dataset.head()
```

* Create a histogram to visualize the distribution of a numerical variable.


```python
plt.hist(dataset['Electric Range'], bins=10, color='blue', edgecolor='black')
plt.xlabel('Electric Range')
plt.ylabel('Frequency')
plt.title('Histogram of Electric Range')
plt.show()
```

* Plot a bar chart to show the frequency of categories in a categorical variable.


```python
category_counts = dataset['Make'].value_counts()
plt.bar(category_counts.index, category_counts.values, color='green')
plt.xlabel('Make')
plt.ylabel('Frequency')
plt.title('Bar Chart of Make of Electric Vehicles')
plt.xticks(rotation=90)
plt.show()
```

* Create a scatter plot to visualize the relationship between two numerical variables.

```python
plt.scatter(dataset['Make'], dataset['Model Year'], color='red')
plt.xlabel('Model Year')
plt.ylabel('Model')
plt.title('Scatter Plot of Make vs Model Year of Electric Vehicles')
plt.xticks(rotation=90)
plt.show()
```
